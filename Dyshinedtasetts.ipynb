```python
df.fillna(0,inplace=True)
df['Parental_Education_Level'].str.strip()
```

```python
#calculate Percentage of Exam
df['ExamPercentages%']=(df['Exam_Score']/df['Exam_Score'].sum()*100)
df['ExamPercentages%'].sort_values(ascending=False)
```

```python
#calculate Privous Exam Scores
df['PriviousPercentages%']=(df['Previous_Scores']/df['Previous_Scores'].sum()*100)
df['PriviousPercentages%'].sort_values(ascending=False)
```

```python
#Distribution plot for numerical values
numerical_values=df.select_dtypes(include=np.number).columns
plt.figure(figsize=(14,6))
for i,col in enumerate(numerical_values):
  plt.subplot(4,4,i+1)
  sns.histplot(df[col],kde=True,bins=30)
  plt.title(f'Distribution plot for {col}')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
#outlier Detection to using boxplot
#outlier indicates the high standard deviation
plt.figure(figsize=(11,5))
for i,col in enumerate(numerical_values):
  plt.subplot(4,4,i+1)
  sns.boxplot(df[col])
  plt.title(f'Distribution plot for {col}')
plt.tight_layout()
plt.show()
```

```python
df.tail()
```

```python
#count plot to use df categorical shows the categorical values
categorical_cols=['Gender','Distance_from_Home','Parental_Education_Level','Peer_Influence','Internet_Access','Motivation_Level','Extracurricular_Activities','Access_to_Resources','Parental_Involvement']
plt.figure(figsize=(15,10))
for i,col in enumerate(categorical_cols):
  plt.subplot(4,4,i+1)
  sns.countplot(y=df[col],order=df[col].value_counts().index[ :10])
  plt.title(f'Categorical column {col}')
plt.tight_layout()
plt.show()
```

```python
#Relationship of numerical variable
#Correlation to drawn the heatMAp
plt.figure(figsize=(12,8))
corrolation_matrix=df[numerical_values].corr()
sns.heatmap(corrolation_matrix,annot=True,fmt='.2f',cmap='coolwarm',linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

```

```python

#Does Family income corrolated with access to resources or final scores
# One-hot encode the 'Family_Income' column
income_dummies = pd.get_dummies(df['Family_Income'], prefix='Income')

# Concatenate the new columns with the original DataFrame
df_encoded = pd.concat([df, income_dummies], axis=1)

# Calculate the correlation with 'Exam_Score'
correlation = df_encoded[['Income_Low', 'Income_Medium', 'Income_High', 'Exam_Score']].corr()

# Display the correlation of income levels with Exam_Score
print(correlation['Exam_Score'])
```
